```python
#Importing the required libraries

import pandas as pd
import numpy as np
import seaborn as sns

import matplotlib.pyplot as plt
import matplotlib.mlab as mlab
import matplotlib
plt.style.use('ggplot')
from matplotlib.pyplot import figure

%matplotlib inline
matplotlib.rcParams['figure.figsize'] = (12,8)

#Taking the data as input
df = pd.read_csv('D:/Python Data Anlysis/archive (1)/movies.csv')
```

```python
df.shape
```

```python

for col in df.columns:
    pct_missing = np.mean(df[col].isnull())
    print('{} - {}%'.format(col, round(pct_missing*100)))
```

```python
df.isnull().sum()
```

```python
print(df.dtypes)
```

```python
df = df.dropna()
df = df.astype({"budget":"int64","gross":"int64"})
```

```python
df.shape
```

```python
df['Year_correct'] = df['released'].str.extract(pat = '([0-9]{4})').astype(int)
```

```python
df.sort_values(by=['gross'], inplace=False, ascending=False)
```

```python
sns.regplot(x='budget', y='gross', data = df, scatter_kws={"color": "red"}, line_kws={"color" : "blue"})
```

```python
sns.regplot(x='score', y='gross',data=df, scatter_kws={"color": "red"}, line_kws={"color" : "blue"})
```

```python
df.corr(method= 'pearson')
```

```python
correlation_matrix = df.corr(method ='pearson')

sns.heatmap(correlation_matrix, annot=True)

plt.title("Correlation matrix for Numeric Features")

plt.xlabel("Movie features")

plt.ylabel("Movie features")

plt.show()
```

```python
df_numerized = df

for col_name in df_numerized.columns:
    if(df_numerized[col_name].dtype == 'object'):
        df_numerized[col_name] = df_numerized[col_name].astype('category')
        df_numerized[col_name] = df_numerized[col_name].cat.codes
        
df_numerized
```

```python
correlation_matrix2 = df_numerized.corr(method ='pearson')

sns.heatmap(correlation_matrix2, annot=True)

plt.title("Correlation matrix for Numeric Features")

plt.xlabel("Movie features")

plt.ylabel("Movie features")

plt.show()
```

```python
df_numerized.corr()
```

```python
correlation_mat = df_numerized.corr()

corr_pairs = correlation_mat.unstack()

corr_pairs
```

```python
sorted_pairs = corr_pairs.sort_values()

sorted_pairs
```

```python
high_corr = sorted_pairs[(sorted_pairs) > 0.5]

high_corr
```
